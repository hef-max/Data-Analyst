remove missing value/NaN

```
df = pd.util.testing.makeMissingDataframe().reset_index()
df.head()
```

```
df = df.rename(columns={'index':'Z'})
df.head()
```

```
df_backup = df.copy(deep=True)
```

```
df = df.dropna(axis='columns')
df.head()
```

```
df = df_backup.copy(deep=True)
df = df.dropna(axis='rows')
df.head()
```

```
df = df_backup.copy(deep=True)
print(round(df.isna().mean() * 100, 1), "%")
```

menghapus berdasarkan threshold

```
thresh = len(df) * 0.9
df = df.dropna(thresh=thresh, axis='columns')
df.head()
```

```
df = pd.read_csv('/content/titanicfull.csv')
df.columns = ['Pclass', 'Survival status', 'full Name', 'Sex ', ' Age',
              'Sib SP', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']
df_backup = df.copy(deep=True)
df.head()
```

```
from IPython.display import JSON
import pandas as pd

api = JSON('https://api-twitter-trends.herokuapp.com/trends?location=indonesia')
api_json = api.data['data']['trends']
# api_json

datetime = []
time = []
name = []
tweet = []
for i in range(0, len(api_json)):
    for x in range(0 , len(api_json[0]['data'])):      
      datetime.append(api_json[i]['datetime'])
      time.append(api_json[i]['time'])
      name.append(api_json[i]['data'][x]['name'])
      tweet.append(api_json[i]['data'][x]['tweet_count'])

data = {'datetime': datetime,
        'time':time,
        'tweet_name':name,
        'tweet_count': tweet}

df = pd.DataFrame(data)
df_backup = df.copy(deep=True)
df['tweet_count'] = df['tweet_count'].str.replace('K', '')
df.head(10)
```
